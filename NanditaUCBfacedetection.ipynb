```python
!pip install opencv-python-headless
!wget https://github.com/opencv/opencv/raw/master/data/haarcascades/haarcascade_frontalface_default.xml -O haarcascade_frontalface_default.xml
!ls /content
import cv2
from google.colab.patches import cv2_imshow

# Path to the uploaded files
haarcascade_path = '/content/haarcascade_frontalface_default.xml'
image_path = '/content/face.jpeg'

# Load the image
img = cv2.imread(image_path)

# Load the Haarcascade file
cascade_classifier = cv2.CascadeClassifier(haarcascade_path)

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect faces
faces = cascade_classifier.detectMultiScale(gray, minNeighbors=5)

# Draw rectangles around detected faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 255, 0), 2)
    cv2.putText(img, 'Face', (x + 10, y + 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 255), 2)

# Display the image
cv2_imshow(img)
```
